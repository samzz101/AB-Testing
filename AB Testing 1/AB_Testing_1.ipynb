```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("bank-full.csv", sep = ";")
```

```python
data['campaign'].median()
```

```python
median_campaign = data['campaign'].median()
```

```python
data['campaign'].describe()
```

```python
data['segment_campaign'] = data['campaign'].apply(lambda x: 'High' if x > median_campaign else 'Low')
```

```python
data
```

```python
data['subscription_status'] = np.where(data['y'] == 'yes', 1, 0)
```

```python
data
```

```python
data['subscription_status'].dtype
```

```python
data.groupby(['segment_campaign'],as_index=False).count()
```

```python
data_sample = data.groupby(['segment_campaign']).apply(lambda x: x.sample(n=5000,random_state=40))
```

```python
data_sample
```

```python
data_sample['segment_campaign'].value_counts()
```

```python
highercamp = data_sample[data_sample['segment_campaign']=='High']
lowercamp = data_sample[data_sample['segment_campaign']=='Low']
```

```python
print(highercamp['subscription_status'].value_counts())
```

```python
print(lowercamp['subscription_status'].value_counts())
```

```python
success_highcamp = highercamp['subscription_status'].sum()
success_highcamp
```

```python
success_lowcamp = lowercamp['subscription_status'].sum()
success_lowcamp
```

```python
observations_highcamp = highercamp['subscription_status'].count()
observations_highcamp
```

```python
observations_lowcamp = lowercamp['subscription_status'].count()
observations_lowcamp
```

```python
cr_highcamp = success_highcamp / observations_highcamp
cr_highcamp
```

```python
cr_lowcamp = success_lowcamp / observations_lowcamp
cr_lowcamp
```

```python
import matplotlib.pyplot as plt

# Conversion rates and labels
campaign_labels = ['High Campaign', 'Low Campaign']
conversion_rates = [cr_highcamp, cr_lowcamp]

# Plotting
plt.figure(figsize=(10, 7))
bars = plt.bar(campaign_labels, conversion_rates)  # Blue and orange colors

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add data labels on the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2%}', ha='center', va='bottom', fontsize=12, color='black')

# Adding titles and labels
plt.title('Conversion Rate Based on Campaign', fontsize=17)
plt.xlabel('Campaign Segment', fontsize=14)
plt.ylabel('Conversion Rate', fontsize=14)
plt.ylim(0, max(conversion_rates) * 1.2)  # Set y-axis limit to be slightly above the highest conversion rate

# Display plot
plt.show()
```

```python
p_pooled = (success_highcamp + success_lowcamp) / (observations_highcamp + observations_lowcamp)
p_pooled
```

```python
import math
```

```python
standard_error = math.sqrt(p_pooled * (1 - p_pooled) * (1 / observations_highcamp + 1 / observations_lowcamp))
standard_error
```

```python
z_score = (cr_highcamp - cr_lowcamp) / standard_error
z_score
```

```python
from scipy import stats
```

```python
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
p_value
```

```python
if p_value >= 0.05:
    print('Insufficient evidence to reject H0')
else:
    print('Sufficient evidence to reject H0')
```
